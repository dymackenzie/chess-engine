```python
import pandas as pd
import re
import string

li = []
paths = ['chess_openings/a.tsv', 'chess_openings/b.tsv', 'chess_openings/c.tsv', 'chess_openings/d.tsv', 'chess_openings/e.tsv']
for path in paths:
    database = pd.read_csv(path, index_col = None, header = 0, sep = "\t")
    li.append(database)
openings = pd.concat(li, axis = 0, ignore_index = True)
```

```python
all_pgn = pd.Series(openings['pgn'], dtype = pd.StringDtype())
all_pgn
```

```python
def convert_pgn(pgn) -> list:
    list_pgn = re.split(".\\.", pgn)[1:]
    end_moves = []
    entries = []
    for x in list_pgn:
        x = x.strip()
        x = x.split(" ")
        if len(x) < 2:
            return []
        entries.append((x[0][-2:], x[1][-2:]))
    for entry in entries:
        try:
            entry = (convert_to_index(entry[0]), convert_to_index(entry[1]))
        except Exception:
            continue
        end_moves.append(entry)
    return end_moves

def convert_to_index(entry) -> int:
    try:
        return 91 + (string.ascii_lowercase.find(entry[0]) * -10) + (int(entry[1]) * -1)
    except Exception:
        raise Exception
```

```python
result = []
for pgn in all_pgn:
    insert = convert_pgn(pgn)
    if len(insert) != 0 and len(insert) > 4:
        result.append(insert)
result
```
